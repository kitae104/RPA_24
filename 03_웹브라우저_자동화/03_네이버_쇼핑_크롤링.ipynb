### 네이버 쇼핑 크롤링
라이브러리 설치 
- pip install selenium
- pip install webdriver-manager

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

options = Options() # 크롬 옵션 객체 생성
options.add_experimental_option("detach", True) # 크롬 창을 종료해도 프로세스가 종료되지 않도록 설정
service = Service(ChromeDriverManager().install())  # 크롬 드라이버 경로 설정
```

```python
driver = webdriver.Chrome(service=service, options=options) # 크롬 드라이버 객체 생성
# 네이버 쇼핑 페이지로 이동합니다.
driver.get('https://shopping.naver.com/home')

# 브라우저 창을 전체 화면으로 설정
driver.maximize_window()
time.sleep(1)
# 검색 입력 필드를 찾아서 검색어를 입력합니다.
search_input = driver.find_element(By.CLASS_NAME, '_searchInput_search_text_3CUDs')
search_input.send_keys("맥 북")

# 엔터 키를 입력하여 검색을 실행합니다.
search_input.send_keys(Keys.ENTER)

# WebDriver를 종료합니다.
#driver.quit()
```

현재 스크롤된 높이 확인  - 무한 스크롤 처리
- window.scrollY

```python
# 스크롤 전 높이
before_height = driver.execute_script("return window.scrollY")

# 무한 스크롤 
# 스크롤을 내리면서 데이터를 수집하는 방법
while True:
    try:
        # 맨 아래로 스크롤 내리기
        more = driver.find_element(By.CSS_SELECTOR, "body").send_keys(Keys.END)
        
        # 스크롤 후 페이지 로딩 시간 대기 
        time.sleep(1)

        # 스크롤 후 높이
        after_height = driver.execute_script("return window.scrollY")
        if before_height == after_height:
            break

        before_height = after_height

    except:
        break
```

```python
import csv

# 파일 생성
f = open("naver_shopping.csv", "w", encoding="utf-8", newline="")  # newline : 줄바꿈 제거

# 상품 정보 div
items = driver.find_elements(By.CSS_SELECTOR, "div.product_item__MDtDF")
print(len(items))
for item in items:
    # 상품명
    name = item.find_element(By.CSS_SELECTOR, "div.product_title__Mmw2K").text
    # 가격
    price = item.find_element(By.CSS_SELECTOR, "span.price_num__S2p_v").text

    # 링크 - 자식 선택자 사용
    link = item.find_element(By.CSS_SELECTOR, "div.product_title__Mmw2K > a").get_attribute("href")

    
    print(name, price, link)
    csv.writer(f).writerow([name, price, link])

f.close()
```
